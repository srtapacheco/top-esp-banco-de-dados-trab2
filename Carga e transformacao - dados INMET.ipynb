```r
# Carregar o pacote 'readr' do tidyverse
library(readr)
library(dplyr)
```

```r
# Carregar o arquivo CSV com configurações específicas

dados <- read_delim("INMET_SE_RJ_A602_RIO DE JANEIRO-MARAMBAIA_01-01-2023_A_31-12-2023.CSV",
                  skip = 8,
                  delim = ";",
                  locale = locale(encoding = "latin1"))
```

```r
# Número de linhas
num_linhas <- nrow(dados)

# Número de colunas
num_colunas <- ncol(dados)

# Número de células (linhas * colunas)
num_celulas <- num_linhas * num_colunas

# Exibir os resultados
cat("Número de linhas:", num_linhas, "\n")
cat("Número de colunas:", num_colunas, "\n")
cat("Número de células:", num_celulas, "\n")
```

```r
# Definindo um vetor de colunas de interesse
colunas_interesse <- c("PRECIPITAÇÃO TOTAL, HORÁRIO (mm)","PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)","PRESSÃO ATMOSFERICA MAX.NA HORA ANT. (AUT) (mB)","PRESSÃO ATMOSFERICA MIN. NA HORA ANT. (AUT) (mB)","TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)","TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)","TEMPERATURA MÍNIMA NA HORA ANT. (AUT) (°C)","VENTO, DIREÇÃO HORARIA (gr) (° (gr))","VENTO, RAJADA MAXIMA (m/s)","VENTO, VELOCIDADE HORARIA (m/s)")

# Filtrando o dataset usando o vetor
dados_filtrados <- dados %>%
  select(all_of(colunas_interesse))
```

```r
# Exibir as primeiras linhas do dataset filtrado
head(dados_filtrados)
```

```r
num_linhas <- nrow(dados_filtrados)
num_colunas <- ncol(dados_filtrados)
num_celulas <- num_linhas * num_colunas

cat("Número de linhas:", num_linhas, "\n")
cat("Número de colunas:", num_colunas, "\n")
cat("Número de células:", num_celulas, "\n")
```

```r
# Filtrar as linhas sem NAs
dados_validos <- dados_filtrados %>%
  filter(complete.cases(.))
```

```r
head(dados_validos)
```

```r
num_linhas <- nrow(dados_validos)
num_colunas <- ncol(dados_validos)
num_celulas <- num_linhas * num_colunas

cat("Número de linhas:", num_linhas, "\n")
cat("Número de colunas:", num_colunas, "\n")
cat("Número de células:", num_celulas, "\n")
```

```r
# Salvar com ponto e vírgula como delimitador
write_delim(dados_validos, "dados_validos2023.csv", delim = ";")
```
